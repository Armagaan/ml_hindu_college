# Penguins
Exploring Continuous Data

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

sns.set()
sns.set_style("white")
plt.style.use("seaborn-v0_8-paper")
plt.rcParams["font.family"] = "monospace"
plt.rcParams['xtick.labelsize'] = 11
plt.rcParams['ytick.labelsize'] = 11
```

```python
df_penguins = pd.read_csv("../data/penguins_size.csv")
df_penguins.head()
```

# Can we visualize any clusters?

```python
fig, ax = plt.subplots()
sns.scatterplot(
    ax=ax,
    data=df_penguins,
    x="culmen_length_mm",
    y="culmen_depth_mm",
    hue="species",      # * Add a dimension as color.
    size="body_mass_g", # * Add a dimension as size.
    style="island"      # * Add a dimension as marker shape.
)
ax.legend(bbox_to_anchor=(1, 1))
fig.savefig("../plots/penguins.png", dpi=200, bbox_inches="tight")
```

```python

# * layout="constrained" manages overlapping parts of the plot.
fig, axes = plt.subplots(1, 2, figsize=(12, 4), layout="constrained")

sns.scatterplot(
    ax=axes[0],
    data=df_penguins,
    x="culmen_length_mm",
    y="culmen_depth_mm",
    hue="species",
    size="body_mass_g",
    alpha=0.75 # * Use alpha to control opacity.
)
sns.scatterplot(
    ax=axes[1],
    data=df_penguins,
    x="culmen_length_mm",
    y="culmen_depth_mm",
    hue="island",
    size="body_mass_g",
    palette="rocket",
    alpha=0.75
)

# * Use bbox_to_achor to set legend position
# * Use frameon to remove legend's frame
axes[0].legend(bbox_to_anchor=(1, 1), frameon=False)
axes[1].legend(bbox_to_anchor=(1, 1))

fig.savefig("../plots/penguins.png", dpi=200, bbox_inches="tight") # * Use dpi and bbox_inches
```

```python

# * Visualize data distribution of the clusters as well.
# We can't pass ax=... here as this is a "figure-level" plot.
sns.jointplot(
    kind="scatter",
    data=df_penguins,
    x="culmen_length_mm",
    y="culmen_depth_mm",
    hue="species",
)
plt.show()
```
