# Conditional indexing

```python
df[df.Age > 30]
```

```python
# Not isna()
~ df.Age.isna()
```

```python
# Filter rows where Age is not NA
df[df.Age.notna()]
```

```python
# Filter rows where Age is not NA
df[ ~ df.Age.isna() ]
```

```python
# Filter rows where Age is between 30 and 40

# Chain conditions using boolean logic.
# Remeber to put each condition within parenthesis.
df[(df.Age < 40) & (df.Age > 30)]
```

```python
# Filter rows where Age is less than 30 or Passenger is a female.
df[(df.Age < 30) | (df.Sex == "female")]
```
